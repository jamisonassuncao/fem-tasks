```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1001)
T = 40.+(66.*x)-(5*x**2)
t = 16.*x+40.+(400./np.pi**2)*np.sin(np.pi*x/10.)+(400./(9.*np.pi**2))*np.sin(3.*np.pi*x/10)
```

```python
lettersize = 20
plt.rcParams.update({'font.size': lettersize})

plt.figure(figsize=(15,10))
plt.plot(x,T,label="Analytical Solution")
plt.plot(x,t,label="Numerical Solution")
plt.xlabel("x")
plt.ylabel("T")
plt.legend()
plt.tight_layout()

plt.savefig("ex1.pdf")
```

```python
x = np.linspace(0,2,1001)
y = x + 3*np.sin(x)/np.sin(2)
y1 = 4.52622*x-0.112931*x**2-0.45009*x**3
# y2 = 3.07058*x-0.0588193*x**2-0.113236*x**3
# y2 = 4.28204*x-0.0932333*x**2-0.179488*x**3
y2 = 6.65809*x-0.217544*(0.5*(-1.+(3.*x**2)))-0.418805*(0.5*((5*x**3)-3*x))
y2 = 4.28204*x-0.0932333*(0.5*(-1.+3.*x**2))-0.179488*(0.5*((5*x**3)-3*x))

plt.figure(figsize=(15,10))
plt.plot(x,y,label="y(x)")
plt.plot(x,y1,label=r"$\tilde{y}_1(x)$")
plt.plot(x,y2,label=r"$\tilde{y}_2(x)$")
plt.xlabel("x")
plt.ylabel("T")
plt.legend()
plt.tight_layout()

plt.savefig("ex2.pdf")
```

```python
x = np.linspace(0,np.pi,1001)

aux1 = (np.exp(np.pi)*np.exp(-x))/(2*(np.exp(np.pi)-1))
aux2 = (np.exp(x))/(2*(np.exp(np.pi)-1))
aux3 = (1./2.)*np.cos(x)
y = aux1 - aux2 - aux3
# y1 = -(np.pi/2.)*np.cos(x)-4.*np.sin(2*x)/3.
y1 = np.cos(x)/2.

plt.figure(figsize=(15,10))
plt.plot(x,y,label="y(x)")
plt.plot(x,y1,label=r"$\tilde{y}(x)$")
plt.hlines(x[0],x[-1],0,linestyles="dotted")
plt.legend()
plt.tight_layout()

plt.savefig("ex4.pdf")
```
